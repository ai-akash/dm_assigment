```python
X=data.drop("OUTCOME",axis=1)
y=data["OUTCOME"]
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,test_size=.25)
```

```python
X_train["CREDIT_SCORE"] = X_train["CREDIT_SCORE"].fillna(X_train["CREDIT_SCORE"].median())
X_test["CREDIT_SCORE"] = X_test["CREDIT_SCORE"].fillna(X_train["CREDIT_SCORE"].median())
X_train["ANNUAL_MILEAGE"] = X_train["ANNUAL_MILEAGE"].fillna(X_train["ANNUAL_MILEAGE"].median())
X_test["ANNUAL_MILEAGE"] = X_test["ANNUAL_MILEAGE"].fillna(X_train["ANNUAL_MILEAGE"].median())
```
